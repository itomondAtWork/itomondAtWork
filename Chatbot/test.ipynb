```python
import requests
import os

# 環境変数の設定
env_vars = {
    "AICORE_AUTH_URL": "https://eu1.authentication.eu10.hana.ondemand.com",
    "AICORE_CLIENT_ID": "sb-1fde05f0-e4ab-4be0-b1e4-bdf08f499679!b4790|aicore!b540",
    "AICORE_CLIENT_SECRET": "bf578406-2429-469b-a20e-f5a6d3372dbc$JG38YIuojBsYYoXp8E4zu0ajux3ZJXY4GG4w0RpGh1Y=",
    "AICORE_BASE_URL": "https://api.ai.prod.eu-central-1.aws.ml.hana.ondemand.com/v2",
    'AICORE_RESOURCE_GROUP': "itomond-resource-group",
}

# 環境変数を設定
for key, value in env_vars.items():
    os.environ[key] = value

# 認証トークンを取得する関数
def get_token():
    auth_url = os.getenv('AICORE_AUTH_URL')
    client_id = os.getenv('AICORE_CLIENT_ID')
    client_secret = os.getenv('AICORE_CLIENT_SECRET')
    
    if not all([auth_url, client_id, client_secret]):
        raise ValueError("環境変数が正しく設定されていません。")
    
    token_url = f"{auth_url}/oauth/token"
    data = {
        'grant_type': 'client_credentials',
        'client_id': client_id,
        'client_secret': client_secret
    }
    
    response = requests.post(token_url, data=data)
    
    try:
        response.raise_for_status()
    except requests.exceptions.HTTPError as err:
        print(f'HTTP error occurred: {err}')  # HTTP エラーメッセージ
        print(f'Response content: {response.content}')  # レスポンスの内容
        return None
    
    token = response.json().get('access_token')
    if not token:
        raise ValueError("トークンが取得できませんでした。")
    
    return token

# トークンを取得
token = get_token()
print(f'Token: {token}')
```

```python
import requests
import os
import json

# 認証情報
auth_url = os.getenv('AICORE_AUTH_URL')
client_id = os.getenv('AICORE_CLIENT_ID')
client_secret = os.getenv('AICORE_CLIENT_SECRET')
base_url = os.getenv('AICORE_BASE_URL')
resource_group = os.getenv('AICORE_RESOURCE_GROUP')

# トークンを取得する関数
def get_token(auth_url, client_id, client_secret):
    data = {
        'grant_type': 'client_credentials',
        'client_id': client_id,
        'client_secret': client_secret
    }
    response = requests.post(auth_url + '/oauth/token', data=data)
    response.raise_for_status()
    return response.json()['access_token']

# トークンを取得
token = get_token(auth_url, client_id, client_secret)

# 設定の作成
headers = {
    'Authorization': f'Bearer {token}',
    'Content-Type': 'application/json',
    'Content-Type': 'application/json',
    'Content-Type': 'application/json',
    'AI-Resource-Group': resource_group  # AI-Resource-Group ヘッダーを追加
}

# 設定の作成リクエストデータ
configuration_data = {
    "name": "gpt-4-32k-itomond",
    "executableId": "azure-openai",
    "scenarioId": "foundation-models",
    "versionId": "0.0.1",
    "parameterBindings": [
        {"key": "modelName", "value": "gpt-4-32k"},
        {"key": "modelVersion", "value": "latest"}
    ],
    "inputArtifactBindings": []
}

# 設定の作成リクエスト
response_config = requests.post(base_url + '/v2/lm/configurations', headers=headers, data=json.dumps(configuration_data))

# エラーの詳細を表示
try:
    response_config.raise_for_status()
except requests.exceptions.HTTPError as err:
    print(f'HTTP error occurred: {err}')  # HTTP エラーメッセージ
    print(f'Response content: {response_config.content}')  # レスポンスの内容

# レスポンスの表示
configuration = response_config.json()
print(configuration)

# デプロイメントの作成リクエストデータ
deployment_data = {
    "configurationId": configuration['id']  # 取得した configurationId を使用
}

# デプロイメントの作成リクエスト
response_deployment = requests.post(base_url + '/v2/lm/deployments', headers=headers, data=json.dumps(deployment_data))

# エラーの詳細を表示
try:
    response_deployment.raise_for_status()
except requests.exceptions.HTTPError as err:
    print(f'HTTP error occurred: {err}')  # HTTP エラーメッセージ
    print(f'Response content: {response_deployment.content}')  # レスポンスの内容

# レスポンスの表示
deployment = response_deployment.json()
print(deployment)

# デプロイメントの詳細取得リクエスト
response_deployment_details = requests.get(base_url + '/v2/lm/deployments', headers=headers)

# エラーの詳細を表示
try:
    response_deployment_details.raise_for_status()
except requests.exceptions.HTTPError as err:
    print(f'HTTP error occurred: {err}')  # HTTP エラーメッセージ
    print(f'Response content: {response_deployment_details.content}')  # レスポンスの内容

# レスポンスの表示
deployment_details = response_deployment_details.json()
print(deployment_details)
```
